### Notebook to mine the ArXiV for paper titles from astrophysics journal papers
(https://arxiv.org/archive/astro-ph)

```python
# Mine the arxiv and extract paper titles 
# Save this as a list and the format for a neural network code (LTSD cell code in GDL book, chapter 6). 

import arxiv # library for accessing ArXiV API
import numpy as np
import pylab
import pandas as pd
import re
import string
```
